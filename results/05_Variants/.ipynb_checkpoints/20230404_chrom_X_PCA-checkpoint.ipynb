## Setup figures

```python
# BDD: Boudiediette
# SRM: Sare Mady
# KRK: Kérakounda
# KLD: Kolda

site_colours = {
    'Boudiediette': '#FF0000',
    'Sare Mady': '#008000',
    'Kérakounda': '#00FFFF',
    'Kolda': '#90EE90',
}
```

```python
def plot_pca_coords(coords, model, pc1, pc2, ax, sample_population):
    sns.despine(ax=ax, offset=5)
    x = coords[:, pc1]
    y = coords[:, pc2]
    for pop in sites:
        flt = (sample_population == pop)
        ax.plot(x[flt], y[flt], marker='o', linestyle=' ', color=site_colours[pop], 
                label=pop, markersize=6, mec='k', mew=.5)
    ax.set_xlabel('PC%s (%.1f%%)' % (pc1+1, model.explained_variance_ratio_[pc1]*100))
    ax.set_ylabel('PC%s (%.1f%%)' % (pc2+1, model.explained_variance_ratio_[pc2]*100))
    

def fig_pca(coords, model, title, sample_population=None):
    if sample_population is None:
        sample_population = df_samples.location.values
    # plot coords for PCs 1 vs 2, 3 vs 4
    fig = plt.figure(figsize=(10, 5))
    ax = fig.add_subplot(1, 2, 1)
    plot_pca_coords(coords, model, 0, 1, ax, sample_population)
    ax = fig.add_subplot(1, 2, 2)
    plot_pca_coords(coords, model, 2, 3, ax, sample_population)
    ax.legend(bbox_to_anchor=(1, 1), loc='upper left')
    fig.suptitle(title, y=1.02)
    fig.tight_layout() 
```

### Removing correlated features (LD pruning)
Estimate the linkage disequilibrium parameter r for each pair of variants using the method of Rogers and Huff (2009):
Linkage disequilibrium is often measured by two statistics, D and r, which can be interpreted as the covariance and the correlation between loci and across gametes. r can be expressed as correlations across diploid genotypes, provided that mating is random.

```python
def plot_ld(gn, title):
    m = allel.rogers_huff_r(gn) ** 2
    ax = allel.plot_pairwise_ld(m)
    ax.set_title(title)
```

```python
plot_ld(gn[:1000], 'Figure 1. Pairwise LD (Correlation r between loci).')
```

The darker regions in the plot above indicate pairs of SNPs where genotypes are correlated.
I’m going to randomly choose a subset of these SNPs to work with. This should still reveal the main signals in the data, while making runtime faster.

To remove this remaining correlation, I’m going to locate SNPs that are not correlated with each other, using the *locate_unlinked* function from scikit-allel. This is known as LD pruning, and works by sliding a window along the data, computing pairwise LD between all SNPs within each window, then removing one SNP from each correlated pair.

```python
# size: Window size (number of variants).
# step: Number of variants to advance to the next window.
# threshold: Maximum value of r**2 to include variants.
def ld_prune(gn, size, step, threshold=.1, n_iter=1):
    for i in range(n_iter):
        loc_unlinked = allel.locate_unlinked(gn, size=size, step=step, threshold=threshold)
        n = np.count_nonzero(loc_unlinked)
        n_remove = gn.shape[0] - n
        print('iteration', i+1, 'retaining', n, 'removing', n_remove, 'variants')
        gn = gn.compress(loc_unlinked, axis=0)
    return gn
```

```python
gnu = ld_prune(gn, size=500, step=200, threshold=.1, n_iter=1)
```

```python
plot_ld(gnu[:1000], 'Figure 2. Pairwise LD after LD pruning.')
```

By randomly downsampling SNPs, this should have dealt with much of the correlation between nearby features. Let’s take a look at the first 1000.

The data are relatively small now after LD-pruning, so we can bring the data out of chunked storage and into memory uncompressed, which is necessary for PCA.

```python
gnu = gnu[:]
gnu
```

### PCA via conventional Singular Value Decomposition

Let’s run a conventional PCA analysis of the LD-pruned genotype data.
Scaling method: 
‘patterson’ applies the method of Patterson et al. 2006 [Population Structure and Eigenanalysis](https://journals.plos.org/plosgenetics/article?id=10.1371/journal.pgen.0020190);   
‘standard’ scales to unit variance;   
None centers the data only.

```python
coords1, model1 = allel.pca(gnu, n_components=10, scaler='patterson')
```

```python
fig_pca(coords1, model1, 'Figure 3. Conventional PCA.')
```

```python
fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(121)
ax.plot(model1.explained_variance_ratio_, 'go')
ax.set_xlabel("principal component")
ax.set_ylabel("variance explained")
```
